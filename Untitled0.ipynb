```python
# Select feature
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt


X1 = df[['Annual Income (k$)']]

# Find optimal k using Elbow + Silhouette
inertias = []
sil_scores = []
K = range(2, 11)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X1)
    inertias.append(kmeans.inertia_)
    sil_scores.append(silhouette_score(X1, kmeans.labels_))

# Plot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

ax1.plot(K, inertias, 'bo-')
ax1.set_title('Elbow Method (Income Only)')
ax1.set_xlabel('Number of Clusters (k)')
ax1.set_ylabel('Inertia')
ax1.grid(True)

ax2.plot(K, sil_scores, 'ro-')
ax2.set_title('Silhouette Score (Income Only)')
ax2.set_xlabel('Number of Clusters (k)')
ax2.set_ylabel('Silhouette Score')
ax2.grid(True)

plt.tight_layout()
plt.show()

X2 = df[['Annual Income (k$)', 'Spending Score (1-100)']]

inertias2 = []
sil_scores2 = []

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X2)
    inertias2.append(kmeans.inertia_)
    sil_scores2.append(silhouette_score(X2, kmeans.labels_))

# Plot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

ax1.plot(K, inertias2, 'bo-')
ax1.set_title('Elbow Method (Income + Spending)')
ax1.set_xlabel('k')
ax1.set_ylabel('Inertia')
ax2.grid(True)

ax2.plot(K, sil_scores2, 'ro-')
ax2.set_title('Silhouette Score (Income + Spending)')
ax2.set_xlabel('k')
ax2.set_ylabel('Score')
ax2.grid(True)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

# Perform 1D Clustering
X1 = df[['Annual Income (k$)']]
kmeans_1d = KMeans(n_clusters=3, random_state=42, n_init=10)
df['Cluster_1D'] = kmeans_1d.fit_predict(X1)

# Perform 2D Clustering
X2 = df[['Annual Income (k$)', 'Spending Score (1-100)']]
kmeans_2d = KMeans(n_clusters=5, random_state=42, n_init=10)
df['Cluster_2D'] = kmeans_2d.fit_predict(X2)

plt.figure(figsize=(16, 7))

# 1D Clustering Visualization
plt.subplot(1, 2, 1)
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)',
                hue='Cluster_1D', palette='deep', s=70, alpha=0.8)
plt.title('K-Means: 1 Feature (Income) → k=3', fontsize=14, fontweight='bold')
plt.legend(title='Cluster', loc='upper right')

# 2D Clustering Visualization
plt.subplot(1, 2, 2)
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)',
                hue='Cluster_2D', palette='Set1', s=70, alpha=0.8)
plt.title('K-Means: 2 Features (Income + Spending) → k=5', fontsize=14, fontweight='bold')
plt.legend(title='Cluster', loc='upper right')

plt.tight_layout()
plt.show()
```

```python
print("1D Clustering - Cluster Means:")
print(df.groupby('Cluster_1D')[['Annual Income (k$)']].mean())

print("\n2D Clustering - Cluster Means:")
print(df.groupby('Cluster_2D')[['Annual Income (k$)', 'Spending Score (1-100)']].mean())
```

Five

```python
# Add cluster labels
cluster_labels = {
    0: "High Income, High Spend (VIP)",
    1: "High Income, Low Spend (Frugal Rich)",
    2: "Medium Income, Medium Spend (Core Shoppers)",
    3: "Low Income, Low Spend (Budget Browsers)",
    4: "Low Income, High Spend (Impulse Buyers)"
}

df['Segment'] = df['Cluster_2D'].map(cluster_labels)

# Summary table
summary = df.groupby('Segment').agg({
    'Annual Income (k$)': 'mean',
    'Spending Score (1-100)': 'mean',
    'CustomerID': 'count'
}).round(1).rename(columns={'Annual Income (k$)': 'Income', 'Spending Score (1-100)': 'SpendingScore', 'CustomerID': 'Size'})

summary = summary.sort_values(['Income', 'SpendingScore'], ascending=[False, False])
print("Customer Segments Summary:")
display(summary)
```
